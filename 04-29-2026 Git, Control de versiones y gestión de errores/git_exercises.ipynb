# 🛠️ Laboratorio práctico de Git

**Instrucciones**: Abre el **Terminal** en VS Code para ejecutar estos comandos. No los ejecutes en celdas de Python.

---

## Parte 1: Configuración inicial

1. Crea una nueva carpeta en tu **Escritorio** llamada `learning-git`.
2. Abre esa carpeta en VS Code.
3. Abre la terminal (`Ctrl + ` ` ` or `Ctrl + ñ`).
4. Inicializa el repositorio:
   ```bash
   git init
   ```
5. Comprueba que se ha creado una carpeta oculta `.git`  (usa `ls -a`).

## Parte 2: El ciclo de commits

1. Crea un archivo llamado `main.py` y escribe `print("Hello Git")` dentro.
2. Comprueba el estado:
   ```bash
   git status
   ```
   *(Deberías ver main.py en rojo — sin seguimiento)*

3. Añade el archivo al área de preparación:
   ```bash
   git add main.py
   ```

4. Comprueba el estado de nuevo (`git status`). Deberia aparecer en verde.

5. Crea el commit:
   ```bash
   git commit -m "Initial commit: created main script"
   ```

6. Comprueba el historial:
   ```bash
   git log
   ```

## Parte 3: Realizar cambios

1. Modifica `main.py`. Añade una segunda linea `print("This is a change")`.
2. Crea un nuevo archivo `utils.py`.
3. Ejecuta `git status`. Verás un archivo **modificado** y un archivo **sin seguimiento**
4. Añade todo de una vez:
   ```bash
   git add .
   ```
5. Haz el commit:
   ```bash
   git commit -m "Added utils and updated main"
   ```

## Parte 4: El archivo `.gitignore`

1. Crea un archivo llamado `secret_passwords.txt`.
2. No queremos esto en Git. Crea un archivo llamado `.gitignore`.
3. Dentro de `.gitignore`, escribe:
   ```text
   *.txt
   ```
4. Ejecuta `git status`. `secret_passwords.txt` **NO** debería aparecer. Solo debería aparecer `.gitignore`.
5. Haz commit del archivo `.gitignore`.

## Parte 5: GitHub (La nube)

1. Ve a GitHub.com y crea un **New Repository** (vacío, sin README).
2. Copia la URL (HTTPS).
3. Vincula tu repo local a GitHub:
   ```bash
   git remote add origin <PASTE_YOUR_URL_HERE>
   ```
4. Renombra tu rama a `main` (si no lo esta):
   ```bash
   git branch -M main
   ```
5. Sube tu código:
   ```bash
   git push -u origin main
   ```
6. Actualiza la página de GitHub. ¡Tu código debería estar ahí!
