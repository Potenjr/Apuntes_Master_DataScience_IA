## 1. Introducción a POO

La programación orientada a objetos organiza el código alrededor de **objetos**.

### Conceptos clave
- **Clase**: la plantilla.
- **Objeto**: una instancia concreta de esa plantilla.
- **Atributo**: la información del objeto.
- **Método**: lo que el objeto sabe hacer.

Piensa así:
- la clase describe,
- el objeto existe,
- los atributos guardan estado,
- los métodos expresan comportamiento.


```python
# Example 1: first class and first object
class Perro:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def ladrar(self, sonido="Guau"):
        print(f"{self.nombre} dice: {sonido}")


mi_perro = Perro("Fido", 3)
print(f"El perro se llama {mi_perro.nombre} y tiene {mi_perro.edad} años.")
mi_perro.ladrar()
```

## 2. Atributos, métodos y `__init__`

El método `__init__` se ejecuta automáticamente cuando creamos un objeto.

Sirve para inicializar sus atributos.

Los métodos usan `self` para acceder a los datos del propio objeto.


```python
import math


class Circulo:
    def __init__(self, radio):
        self.radio = float(radio)

    def area(self):
        return math.pi * (self.radio ** 2)


mi_circulo = Circulo(6)
print("Radio:", mi_circulo.radio)
print("Área:", mi_circulo.area())
```

```python
# Exercise 2

# Your code here

# Definición de la clase
class Rectangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura


# Crear una instancia
mi_rectangulo = Rectangulo(5, 3)

# Calcular el área
resultado = mi_rectangulo.area()
print("El área es:", resultado)
```

## 3. Herencia

La herencia permite reutilizar código y crear jerarquías.

Una clase hija puede:
- heredar atributos y métodos,
- añadir atributos nuevos,
- sobrescribir métodos.

`super()` permite aprovechar el constructor o comportamiento de la clase padre.


```python
class Animal:
    def __init__(self, nombre, patas=4):
        self.nombre = nombre
        self.patas = patas

    def hablar(self):
        print(f"{self.nombre} hace un sonido.")


class Perro(Animal):
    def __init__(self, nombre, raza):
        super().__init__(nombre)
        self.raza = raza

    def hablar(self):
        print(f"{self.nombre} dice: ¡Guau!")


class Gato(Animal):
    def __init__(self, nombre, color):
        super().__init__(nombre)
        self.color = color

    def hablar(self):
        print(f"{self.nombre} dice: Miau")


perro = Perro("Rex", "Labrador")
gato = Gato("Michi", "Blanco")
perro.hablar()
gato.hablar()
```

### Ejercicio 3: `Vehiculo` y `Coche`

1. Crea una clase `Vehiculo` con:
   - atributos: `marca`, `modelo`
   - método: `descripcion()`

2. Crea una clase `Coche` que herede de `Vehiculo` y añada:
   - atributo: `puertas`
   - sobrescritura de `descripcion()` para incluir el número de puertas


```python
# Exercise 3

# Your code here
class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        return f"Vehículo: {self.marca} {self.modelo}"


# Clase derivada
class Coche(Vehiculo):
    def __init__(self, marca, modelo, puertas):
        super().__init__(marca, modelo)  # Llamamos al constructor de la clase padre
        self.puertas = puertas

    def descripcion(self):
        return f"Coche: {self.marca} {self.modelo}, {self.puertas} puertas"


# Crear instancia
mi_coche = Coche("Toyota", "Corolla", 5)

# Llamar al método
print(mi_coche.descripcion())
```

## 4. Métodos especiales: `__str__`

Los métodos especiales permiten integrar mejor nuestros objetos con Python.

Hoy nos quedamos con uno muy útil:

- `__str__`: define cómo se ve el objeto cuando usamos `print()`.

Hace que nuestros objetos sean más legibles y más fáciles de depurar.


```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"{self.nombre}, {self.edad} años"

    def descripcion(self):
        return f"Persona: {self.nombre}, Edad: {self.edad}"


persona = Persona("Lucía", 28)
print(persona)
print(persona.descripcion())
```

## 5. Ejercicio final: sistema de gestión de estudiantes

Ahora vamos a combinar todo.

Crea las siguientes clases:

### `Persona`
- atributos: `nombre`, `edad`
- método: `__str__`

### `Estudiante` (hereda de `Persona`)
- atributo extra: `carrera`
- atributo extra: `notas` (lista)
- método: `promedio()`
- `__str__` personalizado

### `Profesor` (hereda de `Persona`)
- atributo extra: `departamento`
- atributo extra: `asignaturas` (lista)
- método: `agregar_asignatura(nombre)`
- `__str__` personalizado

### Qué hacer después
- crea al menos dos estudiantes con notas
- crea un profesor
- añade asignaturas al profesor
- imprime toda la información


```python
# Final exercise

# Your code here

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"{self.nombre}, {self.edad} años"


# Clase Estudiante
class Estudiante(Persona):
    def __init__(self, nombre, edad, carrera, notas):
        super().__init__(nombre, edad)
        self.carrera = carrera
        self.notas = notas

    def promedio(self):
        if not self.notas:
            return 0
        return sum(self.notas) / len(self.notas)

    def __str__(self):
        return (f"Estudiante: {self.nombre} | Edad: {self.edad} | "
                f"Carrera: {self.carrera} | Promedio: {self.promedio():.2f}")


# Clase Profesor
class Profesor(Persona):
    def __init__(self, nombre, edad, departamento):
        super().__init__(nombre, edad)
        self.departamento = departamento
        self.asignaturas = []

    def agregar_asignatura(self, nombre):
        self.asignaturas.append(nombre)

    def __str__(self):
        asignaturas_str = ", ".join(self.asignaturas) if self.asignaturas else "Ninguna"
        return (f"Profesor: {self.nombre} | Edad: {self.edad} | "
                f"Departamento: {self.departamento} | "
                f"Asignaturas: {asignaturas_str}")


# Crear estudiantes
est1 = Estudiante("Ana", 21, "Informática", [8, 7.5, 9])
est2 = Estudiante("Luis", 23, "Matemáticas", [6, 7, 6.5])

# Crear profesor
prof = Profesor("Carlos", 45, "Ciencias")

# Añadir asignaturas
prof.agregar_asignatura("Programación")
prof.agregar_asignatura("Álgebra")

# Imprimir información
print(est1)
print(est2)
print(prof)
```
