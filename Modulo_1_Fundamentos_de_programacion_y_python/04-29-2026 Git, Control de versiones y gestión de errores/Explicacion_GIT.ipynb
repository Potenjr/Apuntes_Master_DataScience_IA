# 🐙 Git y Control de Versiones: La Guía Definitiva

## 1. ¿Qué es Git?
Git es un **sistema de control de versiones**. Te permite rastrear los cambios en tus archivos, viajar en el tiempo a versiones anteriores y trabajar con otras personas sin sobrescribir el código de los demás.

Piensa en Git como una **máquina del tiempo** para tu proyecto.

---

## Primera configuración (solo una vez)
Antes de usar Git, debes presentarte. Git imprime esta información en cada "guardado" (commit) que realices.

Abre tu terminal (Git Bash en Windows) y ejecuta:

```bash
git config --global user.name "Your Name"
git config --global user.email "your.email@example.com"
```

Revisa tu configuración:
```bash
git config --list
```

## 3. El flujo de trabajo local (Las 3 etapas)

Git tiene tres áreas principales. Entender esto es la clave para dominar Git.

1.  **Directorio de trabajo:** Donde estás escribiendo código actualmente (tus archivos).
2.  **Área de preparación (Staging Area):** Una sala de espera. Eliges qué archivos quieres guardar.
3.  **Repositorio** La base de datos permanente de versiones (commits).

### Diagram:
```text
Directorio de trabajo  --[git add]-->  Área de preparación  --[git commit]-->  Repositorio```

### Comandos clave

| Commando | Descripción |
| :--- | :--- |
| `git init` | Convierte la carpeta actual en un repositorio de Git.|
| `git status` | **Crucial.** Te dice qué está pasando (archivos modificados, archivos en preparación). |
| `git add filename` | Mueve un archivo al Área de preparación. |
| `git add .` | Mueve **TODOS** los archivos modificados al Área de preparación. |
| `git commit -m "msg"` | Guarda una instantánea del Área de preparación de forma permanente. |
| `git log` | Muestra el historial de commits (la línea de tiempo).|
| `git restore filename` | Restaura el archivo al último commit. |


---

## 4. Ramas (Universos paralelos)

Las ramas te permiten trabajar en nuevas funcionalidades sin romper el código principal. Idealmente, nunca trabajes directamente en `main`.



| Comando | Descripción |
| :--- | :--- |
| `git branch` | Lista todas las ramas locales. |
| `git branch new-feature` | Crea una nueva rama llamada `new-feature`. |
| `git switch new-feature` | Cambia a esa rama (en versiones antiguas de Git se usaba `checkout`). |
| `git merge new-feature` | Fusiona `new-feature` en la rama actual. |
| `git branch -d <branch-name>` | Elimina una rama local (seguro; se niega si la rama tiene cambios no fusionados). |
| `git branch -D <branch-name>` | Elimina forzosamente una rama local (usar con precaución; descarta trabajo no fusionado). |
| `git push origin --delete <branch-name>` |Elimina una rama en el remoto (GitHub). |
| `git push origin :<branch-name>` | Alternativa antigua para eliminar una rama remota. |

---

## 5. Repositorios remotos (GitHub/GitLab)

Para compartir tu código o tener una copia de seguridad en la nube.

| Comando | Descripción |
| :--- | :--- |
| `git clone <url>` | Descarga un repositorio desde Internet a tu PC.|
| `git remote add origin <url>` | Vincula tu carpeta local a un repositorio de GitHub. |
| `git push origin main` | Sube tus commits a la nube.|
| `git pull origin main` | Descarga nuevos commits desde la nube. |

---

## 6. El archivo `.gitignore`

Hay archivos que **nunca** queremos subir a Git:
* Contraseñas / claves API.
* Archivos temporales(`__pycache__`, `.DS_Store`).
* Entornos virtuales (`.venv`, `env`).

Para ignorarlos, crea un archivo llamado `.gitignore` y escribe dentro los nombres de los archivos/carpetas.
