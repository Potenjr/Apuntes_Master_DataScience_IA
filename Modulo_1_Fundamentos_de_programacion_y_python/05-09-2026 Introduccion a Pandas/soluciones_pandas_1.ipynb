## 1️⃣ Series: la estructura más simple

Una **Series** es un array unidimensional con índice.

Puedes verla como una columna suelta con etiquetas.


```python
values = [10, 20, 30, 40]
s_list = pd.Series(values, index=['a', 'b', 'c', 'd'])

data_dict = {'Apple': 100, 'Banana': 200, 'Cherry': 150}
s_dict = pd.Series(data_dict)

print(f"Series from list:\n{s_list}\n")
print(f"Series from dict:\n{s_dict}\n")
print(f"Index: {s_dict.index}")
print(f"Values: {s_dict.values}")
```

## 2️⃣ DataFrames: la tabla de datos

Un **DataFrame** es una estructura bidimensional con filas y columnas.

Normalmente se crea a partir de un diccionario de listas.


```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Paris', 'London', 'Tokyo'],
    'Salary': [50000, 60000, 70000, 80000],
}

df_manual = pd.DataFrame(data)
display(df_manual)
print("Shape:", df_manual.shape)
```

## 5️⃣ Selección e indexación

Aquí está uno de los núcleos de la clase:
- `df['columna']` para columnas,
- `df[['col1', 'col2']]` para varias columnas,
- `.iloc` por posición,
- `.loc` por etiqueta.


```python
col_a = df_large['A']
cols_subset = df_large[['A', 'C', 'E']]

display(cols_subset.head(3))
display(df_large.iloc[0:5, 0:3])
display(df_large.loc[0:5, ['A', 'D']])
print("Value at iloc[10, 2]:", df_large.iloc[10, 2])
print("Value at loc[0, 'A']:", df_large.loc[0, 'A'])
```
